```python
iron_man_color = "白色"
```

```python
def iron_man_def(iron_man_color):
    if iron_man_color == "紅色":
        return("鋼鐵⼈盔甲是紅⾊的")
    elif iron_man_color == "綠⾊":
        return("鋼鐵⼈其實也可以參考浩克的顏色")
    elif iron_man_color == "藍⾊":
        return("鋼鐵⼈變成藍⾊⼩精靈也不錯")
    else:
        return("好像沒這種顏⾊",iron_man_color)
```

```python
iron_man_def(iron_man_color)
```
